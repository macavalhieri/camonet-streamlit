```python
df_att = df.groupby('cod_atendimento').agg(
    data_atendimento=('data_atendimento', 'min'),
    cod_paciente=('cod_paciente', 'first'),
    sexo=('sexo', 'first'),
    idade=('idade', 'first'),
    faixa_etaria=('faixa_etaria', 'first'),
    cod_unidade_saude=('cod_unidade_saude', 'first'),
    nome_unidade=('nome_unidade', 'first'),
    especialidade=('especialidade', 'first'),

    # flags clínicas no nível do atendimento
    tem_cid_infeccioso=('e_diag_infeccioso', 'max'),
    tem_antibiotico=('e_antibiotico', 'max'),
    tem_presc_inadequada=('e_presc_inadequada', 'max'),

    # volume
    n_prescricoes=('cod_medicamento', 'count'),
    n_antibioticos=('e_antibiotico', 'sum')
).reset_index()

df_att
```

```python
import pandas as pd
import plotly.express as px

tmp = df_att.copy()

tmp['tem_cid_infeccioso'] = tmp['tem_cid_infeccioso'].fillna(0).astype(int)
tmp['tem_antibiotico'] = tmp['tem_antibiotico'].fillna(0).astype(int)

tmp['CID infeccioso'] = tmp['tem_cid_infeccioso'].map({0: 'Não', 1: 'Sim'})
tmp['Recebeu ATB'] = tmp['tem_antibiotico'].map({0: 'Não', 1: 'Sim'})

mat = (
    tmp
    .groupby(['CID infeccioso', 'Recebeu ATB'], dropna=False)
    .size()
    .reset_index(name='atendimentos')
)

mat['CID infeccioso'] = pd.Categorical(mat['CID infeccioso'], categories=['Não', 'Sim'], ordered=True)
mat['Recebeu ATB'] = pd.Categorical(mat['Recebeu ATB'], categories=['Não', 'Sim'], ordered=True)
mat = mat.sort_values(['CID infeccioso', 'Recebeu ATB'])

fig = px.density_heatmap(
    mat,
    x='Recebeu ATB',
    y='CID infeccioso',
    z='atendimentos',
    text_auto=True,
)
fig.update_layout(height=320, margin=dict(l=20, r=20, t=40, b=20))
fig.show()
```
